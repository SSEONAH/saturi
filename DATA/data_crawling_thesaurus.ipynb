## preprocess

```python
def preprocess(local_url, local_name):
    df_text = []
    url = local_url
    page = requests.get(url)
    soup = bs(page.text, "html.parser")
    sentence = soup.section.ul.get_text()
    df_text.append(sentence)
    
    for s in range(2, 5000):
        test_url = f'https://wordrow.kr/%EC%82%AC%ED%88%AC%EB%A6%AC/%EA%B0%95%EC%9B%90%EB%8F%84/%EB%AA%85%EC%82%AC/?%EC%AA%BD={s}'
        page_test = requests.get(test_url)
        soup_test = bs(page_test.text, "html.parser")
        sentence_test = soup_test.section.ul.get_text()
    
        if sentence == sentence_test:
            print("모두 완료 되었습니다. 작업을 종료합니다.")
            global final = '\n\n'.join(df_text)
            break
        else:
            df_text.append(sentence_test)
            
    sentence = final.split('\n\n\n\n\n\n ')
    text = []
    for i in sentence :
        i = re.sub('[\n‘’]','',i)
        i = i.strip()
        x,y = i.split(':')
        x = re.sub('\(.*\)','',x)
        text.append((x,y))
    
    df = pd.DataFrame(text,columns =['dial','meaning'])
    df['original'] = df['meaning'].apply(lambda x : x.split('의 방언, ')[0])
    df['reg'] = f'{local_name}'
    df = df[['dial','original','meaning','reg']]
    return df
```

```python
local_url = 'https://wordrow.kr/%EC%82%AC%ED%88%AC%EB%A6%AC/%EA%B0%95%EC%9B%90%EB%8F%84/%EB%AA%85%EC%82%AC/'
local_name = 'kw'
```

```python
preprocess(local_url, local_name)
```
