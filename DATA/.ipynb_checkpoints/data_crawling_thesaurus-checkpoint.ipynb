```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
kw = 'https://wordrow.kr/%EC%82%AC%ED%88%AC%EB%A6%AC/%EA%B0%95%EC%9B%90%EB%8F%84/%EB%AA%85%EC%82%AC/'
```

```python
url = kw
page = requests.get(url)
```

```python
soup = bs(page.text, "html.parser")
```

```python
print(soup.prettify())
```

```python
import re
a = soup.section.ul.get_text()
```

```python
a
```

```python
a = a.split('\n\n\n\n\n\n ')
text = []
for i in a :
    i = re.sub('[\n\‘\’]','',i)
    i = i.strip()
    x,y = i.split(':')
    text.append((x,y))
text
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(text,columns =['dial','meaning'])
df['original'] = df['meaning'].apply(lambda x : x.split('의 방언, ')[0])
df['reg'] = 'kw'
df = df[['dial','original','meaning','reg']]
```
